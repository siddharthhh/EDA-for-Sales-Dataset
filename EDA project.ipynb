```python
auto_med =data[data['Product_Category']=='Auto']['Amount'].median()
auto_med
```

```python
FS_med =data[data['Product_Category']=='Footwear & Shoes']['Amount'].median()
FS_med
```

```python
food_med =data[data['Product_Category']=='Food']['Amount'].median()
food_med
```

```python
Fur_med =data[data['Product_Category']=='Furniture']['Amount'].mean()
Fur_meds=round(Fur_med,0)
Fur_meds
```

```python
data['Amount'][(data['Product_Category']=='Auto')& (data['Amount'].isna())]=data['Amount'][(data['Product_Category']=='Auto')& (data['Amount'].isna())].fillna(auto_med)
data
```

```python
data['Amount'][(data['Product_Category']=='Food')& (data['Amount'].isna())]=data['Amount'][(data['Product_Category']=='Food')& (data['Amount'].isna())].fillna(food_med)
data
```

```python
data['Amount'][(data['Product_Category']=='Footwear & Shoes')& (data['Amount'].isna())]=data['Amount'][(data['Product_Category']=='Footwear & Shoes')& (data['Amount'].isna())].fillna(food_med)
data
```

```python
data['Amount'][(data['Product_Category']=='Furniture')& (data['Amount'].isna())]=data['Amount'][(data['Product_Category']=='Furniture')& (data['Amount'].isna())].fillna(Fur_meds)
data
```

```python
a =sns.countplot(x='Gender',hue='Gender',data = data)

plt.title('Count of Orders by Gender')
plt.xlabel('Gender')
plt.ylabel('Count of Orders')

for bar in a.containers:
    a.bar_label(bar)
```

```python
amount_gen=data.groupby(['Gender'],as_index= False)['Amount'].meaan().sort_values(by ='Amount',ascending = False)
amount_gen
```

```python
age_Sales_group =data.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by ='Amount')

plt.title('Amount Spent by Age group')
plt.xlabel('Age Group')
plt.ylabel('Amount')

sns.barplot(x='Age Group',y='Amount',data = age_Sales_group,hue ='Age Group' )
```

```python
a =sns.countplot(x='Age Group',hue='Gender',data = data)

plt.title('Count of Orders by Age group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Count of orders')

for bar in a.containers:
    a.bar_label(bar)
```

```python
statewise =data.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by ='Orders',ascending = False).head(10)

plt.figure(figsize=(13, 5))
c=sns.barplot(x='State',y='Orders',hue='State',data= statewise)

plt.title('Count of Orders by State')
plt.xlabel('State')
plt.ylabel('Count of orders')

for bar in c.containers:
    c.bar_label(bar)
```

```python
statewise =data.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by ='Amount',ascending = False).head(10)

plt.figure(figsize=(15, 5))
c=sns.barplot(x='State',y='Amount',hue='State',data= statewise)

plt.title('Amount Spent by State')
plt.xlabel('State')
plt.ylabel('Amount')
```
